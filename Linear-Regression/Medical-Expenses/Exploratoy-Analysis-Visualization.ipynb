## Exploratory Analysisi and Visualization

```python
import plotly.express as px
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
"""Aqui estoy definiendo parametros por defecto para evitar repeticion"""
sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (10, 6)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
medical_df.age.describe()
```

**Uniform Distribution (Ages)**

```python
fig = px.histogram(medical_df, # DataFrame de donde se saca la informacion
                   x='age', # Inidica que la columna 'age' se usara para el eje X
                   marginal='box', # Dibuja grafico marginal en la parte de arriba (otras opciones: 'rug', 'violin', 'histogram')
                   nbins=47, # Define el numero de barras (bins)
                   title='Distribution of Age')
fig.update_layout(bargap=0.1) # bargap = espacio entre barras, 0.1 = 10%
fig.show()
```

**Gaussian Distribution (BMI)**

```python
fig = px.histogram(medical_df, 
                   x='bmi', 
                   marginal='box', 
                   color_discrete_sequence=['red'], # fuerza  color especifico a los bins del hist
                   title='Distribution of BMI (Body Mass Index)')
fig.update_layout(bargap=0.1)
fig.show()
```

**The distribution follows a "power law" (Charges)**

```python
fig = px.histogram(medical_df,
                   x='charges',
                   marginal='box',
                   color='smoker', #Divide el hist en subgrupos basado en la columna 'smoker'
                   color_discrete_sequence=['green', 'grey'],#Al usarlo junto a 'color', asigna colores a las categorias 
                   title='Annual Medical Charges')
fig.update_layout(bargap=0.1)
fig.show()
```

**EXERCISE: Visualize the distribution of medical charges in connection with other factors like "sex" and "region". What do you observe?**

```python
fig = px.histogram(medical_df,
                   x='charges',
                   color='sex',
                   facet_col='region',
                   marginal='box',
                   title='Distribution of charges by sex and region')
fig.update_layout(bargap=0.1)
fig.show()
```
